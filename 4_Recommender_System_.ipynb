```python
# Fill missing values in the 'concern 2' column of DataFrame df1 with an empty string ('').
df1['concern 2'].fillna('', inplace = True)
df1['concern 3'].fillna('', inplace = True)
```

```python
# Concatenate values from columns 'concern', 'concern 2', and 'concern 3' with commas and assign the result to a new column 'concern' in df1.

df1['concern'] = df1['concern'] + ',' + df1['concern 2'] + ',' + df1['concern 3']
```

```python
# Replace occurrences of ' and ' with commas, then replace ' or ' with commas,
# and finally remove any consecutive commas, updating the 'concern' column in df1.

df1['concern'] = df1['concern'].str.replace(' and ', ',').str.replace(' or ', ',').str.replace(",,", "")
```

```python
# Filter DataFrame df1 to include rows where the 'label' column matches any of the specified categories

df = df1[((df1['label'] == 'face-moisturisers') | (df1['label'] == 'mask-and-peel') | (df1['label'] == 'cleanser') | (df1['label'] == 'eye-cream') | (df1['label'] == 'sunscreen'))]
```

```python
#drop the rows which has skin type as null 
df = df[df['skin type'].isna() == False]
```

```python
#rearranging the index values
df.index = [i for i in range(0, len(df))]
df.info()
```

```python
df['concern'] = df['concern'].str.lower()
df['brand'] = df['brand'].str.lower()
df['name'] = df['name'].str.lower()
df['skin type'] = df['skin type'].str.lower()
```

```python
#fill out missing values
top_concerns = {
    'face-moisturisers': 'general care',
    'mask-and-peel': 'general care',
    'cleanser': 'general care',
    'eye-cream': 'fine lines, wrinkles, dark circles, eye bags',
    'sunscreen': 'daily use, general care'
}

# Fill missing concerns based on labels
condition = df['concern'].isnull()  # Filter rows where concern is missing
label_filter = df['label'].map(top_concerns)  # Map labels to top concerns
df.loc[condition, 'concern'] = label_filter  # Replace missing concerns
```

```python
features = ['normal','dry','oily', 'acne','sensitive', 'wrinkles', 'redness', 'dull', 'pore', 'pigmentation', 'blackheads', 'dark circles']
```

```python
# Create a zero-filled NumPy array with dimensions [number of rows in df, number of features].

one_hot_encodings = np.zeros([len(df), len(features)])
```

```python
#skin types first
for i in range(len(df)):
    for j in range(5):
        target = features[j]
        sk_type = df.iloc[i]['skin type']
        if sk_type == 'all':
            one_hot_encodings[i][0:5] = 1
        elif target == sk_type:
            one_hot_encodings[i][j] = 1

#other features
for i in range(len(df)):
    for j in range(5, len(features)):
        feature = features[j]
        if feature in df.iloc[i]['concern']:
            one_hot_encodings[i][j] = 1
```

```python
# Function wrap takes a list info_arr and returns a dictionary with keys 'brand', 'name', 'price', 'url', 'skin type', and 'concern'.

def wrap(info_arr):
    result = {}
    result['brand'] = info_arr[0]
    result['name'] = info_arr[1]
    result['price'] = info_arr[2]
    result['url'] = info_arr[3]
    result['skin type'] = info_arr[4]
    result['concern'] = str(info_arr[5]).split(',')
    return result
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# recommend top 5 similar items from a category
def find_cs(vector=None, label=None, count=5):
    
    # Calculate cosine similarity between vector and one-hot encoded data
    cs_values = cosine_similarity(np.array([vector, ]), one_hot_encodings)
    
    # Assign cosine similarity values to 'cs' column in DataFrame df
    df['cs'] = cs_values[0]
    
    # Filter DataFrame to include items with specified label
    items = df[df['label'] == label]
    
    # Get top recommendations based on cosine similarity
    recommendations = items.sort_values('cs', ascending=False).head(count)
    
    # Convert recommendations DataFrame to list of dictionaries
    data = recommendations[['brand', 'name', 'price', 'url', 'skin type', 'concern']].to_dict('split')['data']
    
    # Wrap each recommendation's information into a dictionary and append to products list
    products = [wrap(element) for element in data]
    
    return products
```

```python
LABELS = list(df.label.unique())
print(LABELS)
```

```python
# overall recommendation
def recommend(vector = None):
    response = {}
    for label in LABELS:
        r = find_cs(vector, label)           
        response[label] = r
    return response
```

```python
#Load the trained skin classification model
import tensorflow as tf

print('Loading model ...')
model = tf.keras.models.load_model('saved_model/mobilenet_model')

class_names = ['Dry_skin', 'Normal_skin', 'Oil_skin']

def load_and_prep_image(filename, img_shape=224):
    img = tf.io.read_file(filename)
    # Decode it into a tensor
    img = tf.image.decode_jpeg(img)
    # Resize the image
    img = tf.image.resize(img, [img_shape, img_shape])
    # Rescale the image (get all values between 0 and 1)
    img = img / 255.
    return img

def predict_class(filename):
    print('Loading image ...')
    # Import the target image and preprocess it
    img = load_and_prep_image(filename)

    print('Predicting class of image ...')

    # Make a prediction
    pred = model.predict(tf.expand_dims(img, axis=0))
    print('Prediction probabilities:', pred)

    # Add in logic for multi-class & get pred_class name
    if len(pred[0]) > 1:
        pred_class = class_names[tf.argmax(pred[0])]
    else:
        pred_class = class_names[int(tf.round(pred[0]))]
    print('Predicted class:', pred_class)
    return pred

skin_type=predict_class(r"C:\Users\SAJJAD-PC\Downloads\Major_Project\Image_data\extra\oily53.jpeg")
```
