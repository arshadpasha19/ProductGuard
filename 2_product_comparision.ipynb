```python
def filter_ingredients(ingredients):
    # Filter rows based on whether the 'IngredientName' is in the ingredient list
    filtered_dataset = df[df['Product Name'].isin(ingredients)]
    
    # Reset the index of the filtered dataset
    filtered_dataset.reset_index(drop=True, inplace=True)
    
    return filtered_dataset


# Call the function to filter the dataset
ingredients1 = filter_ingredients(product1)
ingredients2 = filter_ingredients(product2)
```

```python
# Create a new column 'Safety Level' based on the specified ranges
ingredients1['Safety Level'] = pd.cut(ingredients1['Product Score'], bins=[0, 2, 6, 10], labels=['Safe', 'Moderate', 'Harmful'], right=False)

# Count the occurrences of each safety level
safety_counts = ingredients1['Safety Level'].value_counts()

# Convert the index to strings for concatenation
safety_counts.index = safety_counts.index.astype(str)

# Pie chart
colors = ['#33CC33', '#F4D03F', '#CB4335']
plt.figure(figsize=(8, 8))
plt.pie(safety_counts, labels=safety_counts.index + ' (' + safety_counts.astype(str) + ')', autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Safety Level Distribution of Ingredients - Product 1')
plt.show()
```

```python
def calculate_normalized_safety_score(product):
    total_score = 0
    num_ingredients = len(product)
    for ingredient in product:
        # Calculate score for concerns
        score = 0

        # Include the product score in the calculation
        product_score = ingredient.get("Product Score")
        score += product_score

        # Add points for safety concerns
        score += get_concern_score(ingredient.get("Allergies & Immunotoxicity"))
        score += get_concern_score(ingredient.get("Cancer Concern"))
        score += get_concern_score(ingredient.get("Developmental and Reproductive Toxicity"))
        score += get_concern_score(ingredient.get("Use Restrictions"))

        total_score += score

    # Calculate normalized score per ingredient
    normalized_score = total_score / num_ingredients
    
    return normalized_score


def get_concern_score(concern):
    if concern == "MODERATE":
        return 1
    elif concern == "HIGH":
        return 2
    else:
        return 0
```
